## No more non-numerical values

We will use the RobustScaller because it deals well with outliers, and since it's a large data set, outliers can deal severe damage

```python
y = data['round_winner']
X = data.drop('round_winner', axis = 1)
```

```python
scaler = RobustScaler()
X = scaler.fit_transform(X)
pd.DataFrame(X)
```

```python
sns.heatmap(pd.DataFrame(X).corr())
```

```python
X.shape
```

Using PCA since we can't understand the heatmap due to big numbers of features

```python
pca = PCA(n_components = 96)#number of features
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,10))
plt.hist(pca.explained_variance_ratio_, bins=96)
```

```python
def getKComponents(pca, alpha):
    total_variance = 0
    
    for feature, variance in enumerate(pca.explained_variance_ratio_):
        total_variance += variance
        if (total_variance >= 1- alpha):
            return feature + 1
    return len(pca.explained_variance_ratio_)
```

```python
getKComponents(pca, 0.1) # variance level can't go below 90%
```

the answer is 25 aka we have to add 25 features to maintain 90% variance level

```python
K = getKComponents(pca, 0.05) # variance level can't go below 95%, Expected answer is more than 25
```

```python
X = pca.transform(X)[:, 0:K] ##Transforming X into princible components AKA Better features , but we need only the first K columns
pd.DataFrame(X) ##33 columns --> Dimensionality Reduced
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , train_size = 0.8)
```

```python
log_model = LogisticRegression(verbose=True)
nn_model = MLPClassifier(verbose=True,max_iter = 100)


log_model.fit(X_train,y_train)
nn_model.fit(X_train,y_train)
```

```python
print(f'Log_Model:{log_model.score(X_test,y_test)}')
print(f'NN_Model:{nn_model.score(X_test,y_test)}')
```
